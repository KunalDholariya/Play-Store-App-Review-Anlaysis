```python
# Import Libraries

import pandas as pd
```

```python
# Load dataset

df = pd.read_csv('Play Store Data.csv')
```

```python
# Dataset First Look of play_store data

df.head()
```

```python
# Dataset Last look of play_store data

df.tail()
```

```python
# Dataset Rows & Columns count

print(f'Play Store Dataset has {df.shape[0]} Rows. \nPlay Store Dataset has {df.shape[1]} Columns.')
```

```python
# Play store Dataset Duplicate Value Count

print(f'Play Store Dataset has {df.duplicated().sum()} duplicates values')
```

```python
# Drop duplicates values from the Play Store Dataset

df.drop_duplicates()
```

```python
# Null values from the Play Store Dataset

df.isna().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.isna().sum()
```

```python
# Play Store Dataset Info

df.info()
```

```python
# Changing the 'Reviews' column values into valid numeric values

df['Reviews'] = df['Reviews'].astype(int)
```

```python
# Remove "$" from "Price" columns to make it numeric

df['Price'] = df['Price'].str.replace('$','',regex = False).astype(float)
```

```python
#Changing the 'Last Updated' column values into valid datetime format

df['Last Updated'] = pd.to_datetime(df['Last Updated'])
```

```python
# Remove signs & symbols  from "Price" columns to make it numeric

df['Installs'] = df['Installs'].str.replace(',','').str.replace('+','')
df['Installs'] = df['Installs'].astype(int)
```

```python
# Q1 - Total Number of App Titles Contain Astrology

Astrology_Apps = df[df['App'].str.contains('Astrology',case = False)]
Astrology_Apps
```

```python
# Q2 - Find Average App Rating

avg_app_rating = df['Rating'].mean()
print(f"Average App Ratings :",avg_app_rating)
```

```python
# Q3 - Find Total Number of Unique Category

print(f"Play Store Dataset cointains {df['Category'].nunique()} unique category.\n {'-'*50} \n Category Names : \n",df['Category'].unique())
```

```python
# Q4 - Which Category Has The Highest Average Rating?

df[['Category','Rating']].groupby('Category').mean().sort_values(by = 'Rating' ,ascending = False)
```

```python
# Q5- Find Total Number of App having 5 Star Rating

print(f"Total Apps having 5 Star ratings : {len(df[df['Rating']==5.0])}")
```

```python
# Q6 - Find Average Value of Reviews
 
print(f"Average Ratings Of Apps : {df['Reviews'].mean()}")
```

```python
# Q7 - Find Total Number of Free and Paid Apps

df['Type'].value_counts()
```

```python
# Q8 - Which App Has Maximum Reviews?

df[['App', 'Reviews']].sort_values(by='Reviews', ascending=False).head(1)
```

```python
# Q8 - Display Top 5 Apps Having Highest Reviews

df[['App', 'Reviews']].sort_values(by='Reviews', ascending=False).head(5)
```

```python
# Q9 - Find Average Rating of Free and Paid Apps

df.groupby('Type')['Rating'].mean()
```

```python
# Q10 - Display Top  5 Apps Having Maximum Installs
df[['App','Installs']].sort_values(by= 'Installs', ascending = False).head(5)
```
